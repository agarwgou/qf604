### Simulated Disaster Monthly Frequencies

```python
import numpy as np
import matplotlib.pyplot as plt

lambda_parameter = 30

events = np.random.poisson(lambda_parameter, size=30*12)

plt.hist(events, bins=20, density=True, alpha=0.7, color='blue', edgecolor='black')
plt.title('Poisson Distribution')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.show()
```

```python
date_range = pd.date_range(start='1994-01-01', periods=30*12, freq='M')
sim_disaster_data = pd.DataFrame({'Date': date_range, 'Monthly_Disaster_Freq': events})
sim_disaster_data.set_index('Date', inplace=True)
sim_disaster_data
```

```python
sim_disaster_data.plot(figsize=(18,8))
```

### Calculate and Plot Weight

```python
k = 8 # lags
w2=5
```

```python
def weights (w2,k):
    den = 0
    w =[]
    for i in range(1,k+1):
        den += (1- i/k)**(w2-1)
    
    for j in range(1,k+1):
        num = (1- j/k)**(w2-1)
        w_ = num/den
        w.append(w_)
        
    return w
```

```python
w=weights(w2,k)
```

```python
w
```

```python
plt.plot(range(1,k+1),w)
```

### Realized Average

```python
def sum_of_squares(values):
    return (values ** 2).sum()
```

```python
natgas_futures_data['daily_return']
```

```python
natgas_futures_rv = natgas_futures_data['daily_return'].resample('M').agg(sum_of_squares)
natgas_futures_rv
```

```python
Y= pd.DataFrame(natgas_futures_rv)
Y
```

```python
cpu_data_x = cpu_data['cpu_index'].resample('M').sum()
cpu_data_x
```

```python
X=pd.DataFrame(cpu_data_x)
X
```

```python
import statsmodels.api as sm
```

```python
for k in range(1,k+1):
    X[f'cpu_index_lag{k}'] = X['cpu_index'].shift(k)
    X.dropna(inplace=True)
X.drop(columns=['cpu_index'],inplace=True)
X=X * w
X
```

```python
w
```

```python
Y= Y.tail(X.shape[0])
```

```python
X = sm.add_constant(X)
```

```python
X
```

```python
Y
```

```python
model = sm.OLS(Y, X)
results = model.fit()

# Print regression results
print(results.summary())
```
