```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/forwardcurves.csv')
df.set_index('Contract', inplace = True)
df
```

```python
df.index = pd.to_datetime(df.index, format = "%d/%m/%Y").strftime('%b-%y')
```

```python
df.plot(title = 'Forward curves of Natural Gas, S&P 500, and Bitcoin',
        figsize = (10,5))
plt.ylabel('Normalized price ($), rebased to front month contract')
plt.xlabel('Contract Month')
```
